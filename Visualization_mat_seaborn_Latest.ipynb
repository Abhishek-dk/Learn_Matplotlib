# seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips =sns.load_dataset('tips')
```

```python
tips.head(50)
```

## Univariate  one variable data visualization

```python
#strip plot

sns.stripplot(y= 'tip', data=tips,jitter=True)
plt.ylabel('tip ($)')
plt.show()

#many people give 2-4 so there is cances of "keep change"
```

```python
#Grouping with stripplot()
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()

#weekend(saturday-sunday) more peoples party
```

```python
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Swarm plot

sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
    
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
#Changing orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h')
plt.xlabel('tip ($)')
plt.show()
```

### Box and Violin plot

```python
#plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
```

```python
#plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
plt.subplot(1,2,1) # 1 row 2 coulmn first visual boxplot
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.subplot(1,2,2) # 1 row 2 coulmn first visual violinplot
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')


plt.show()
```

```python
##Combining plots

sns.violinplot(x='day', y='tip', data=tips, inner=None,color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

## Bivariate  two variables data visualization

```python
#Joint plot

sns.jointplot(x= 'total_bill', y= 'tip', data=tips)
plt.show()
```

```python
# Density plot
sns.jointplot(x='total_bill', y= 'tip', data=tips,kind='kde')#kernl density estimation
plt.show()
```

```python
#Pair plot- all types of plot(relation betn 2 values)

sns.pairplot(tips)
plt.show()#all numerica value(total bill,tips,size)
```

```python
sns.pairplot(tips, hue='sex')#hue is used for 3d for 3rd domain in functions like in swarm plt ,pair plt,....
plt.show()
```
